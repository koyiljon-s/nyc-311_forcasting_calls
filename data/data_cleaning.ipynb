```python
# Import Libraties 
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
import numpy as np
```

```python
# Load the Data
df = pd.read_csv('nyc_311_filtered.csv')
df.head(2)
```

```python
# Check missing values 
df.isna().sum()
```

```python
# Drop the columns th
columns_to_drop = [
    'closed_date',
    'descriptor',
    'location_type',
    'incident_zip',
    'incident_address',
    'street_name',
    'cross_street_1',
    'cross_street_2',
    'intersection_street_1',
    'intersection_street_2',
    'address_type',
    'city',
    'landmark',
    'facility_type',
    'due_date',
    'resolution_description',
    'resolution_action_updated_date',
    'bbl',
    'x_coordinate_state_plane',
    'y_coordinate_state_plane',
    'park_facility_name',
    'vehicle_type',
    'taxi_company_borough',
    'taxi_pick_up_location',
    'bridge_highway_name',
    'bridge_highway_direction',
    'road_ramp',
    'bridge_highway_segment',
    'latitude',
    'longitude',
    'location'
]
df = df.drop(columns=columns_to_drop)
```

```python
df.isna().sum()
```

```python

df.sample(3)
```

```python
# Fisrt convert categories to str 
data = df.copy()
data['agency'] = data['agency'].astype(str)
data['agency_name'] = data['agency_name'].astype(str)
data['complaint_type'] = data['complaint_type'].astype(str)
data['status'] = data['status'].astype(str)
data['community_board'] = data['community_board'].astype(str)
data['borough'] = data['borough'].astype(str)
data['open_data_channel_type'] = data['open_data_channel_type'].astype(str)
data['park_borough'] = data['park_borough'].astype(str)
```

```python
# Then, convert them to int
lb = LabelEncoder()

data['agency_name'] = lb.fit_transform(data['agency_name'])
data['complaint_type'] = lb.fit_transform(data['complaint_type'])
data['agency'] = lb.fit_transform(data['agency'])
data['status'] = lb.fit_transform(data['status'])
data['community_board'] = lb.fit_transform(data['community_board'])
data['borough'] = lb.fit_transform(data['borough'])
data['open_data_channel_type'] = lb.fit_transform(data['open_data_channel_type'])
data['park_borough'] = lb.fit_transform(data['park_borough'])
```

```python
data.info()
```

```python
data.sample(2)
```

```python
# Seperate the data into input and output 

# Define date ranges
train_start = pd.Timestamp("2024-08-01")
train_end   = pd.Timestamp("2025-04-30")
test_start  = pd.Timestamp("2025-05-01")
test_end    = pd.Timestamp("2025-08-01")

data['created_date'] = pd.to_datetime(data['created_date'])

df_train = data[(data['created_date'] >= train_start) & (data['created_date'] <= train_end)].copy()
df_test  = data[(data['created_date'] >= test_start)  & (data['created_date'] <= test_end)].copy()

# Save as a csv format
df_train.to_csv("train.csv", index=False)
df_test.to_csv("test.csv", index=False)
```
